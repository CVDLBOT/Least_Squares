## Question 1- part 1-B

```python
# Getting data:
a = .44
b = 1
c = .5
x1 = np.linspace(0, 10, 15)
y = -1*(a/b)*x1 + (-1*(c/b))
y = np.array(y)

print ('Expected, a = 0.44')
print ('Expected, b = 1.0')
print ('Expected, c = 0.5')

data = plt.plot(x1,y, 'ro')
plt.xlabel('x')
plt.ylabel('y = mx + c')
plt.show()
```

```python
# Total Least Squares:
def line_total_least_squares(x,y):
    n = len(x)
    
    x_ = np.sum(x)/n
    y_ = np.sum(y)/n
    
    # Calculate the x~ and y~ 
    x1 = x - x_
    y1 = y - y_
    
    # Create the matrix array
    X = np.vstack((x1, y1))
    X_t = np.transpose(X)
    
    # Finding A_T_A and it's Find smallest eigenvalue::
    prd = np.dot(X,X_t)
    W,V = np.linalg.eig(prd)
    small_eig_index = W.argmin()
    a,b = V[:,small_eig_index] 
    
    # Compute C:
    c = (-1*a*x_) + (-1*b*y_)
    
    return a,b,c
```

```python
# Total Least Squares:
print ('Please wait Until Training is completed!')
print ('Training................................')
a1,b1,c1 = line_total_least_squares(x1,y)
print ('Training Complete')
print 'a = ', a1
print 'b = ', b1
print 'c = ', c1
```

```python
# Plotting:
y_fitted = -1*(a1/b1)*x1 + (-1*(c1/b1))
data = plt.plot(x1,y, 'ro')
fitted = plt.plot(x1,y_fitted)
plt.xlabel('x')
plt.ylabel('y = mx + c')
plt.show()
```
