```python
# So the randomness isn't so random
np.random.seed(0)
```

```python
# Getting data:
m = .44
c = 1
x1 = np.linspace(0, 10, 15)
y = m*x1 + c
y = np.array(y)
print ('Expected, m = 0.44')
print ('Expected, c = 1.0')
data = plt.plot(x1,y, 'ro')
plt.xlabel('x')
plt.ylabel('y = mx + c')
plt.show()
```

```python
# Gradient Descent
def gradientDescent(A,y, x, alpha, numItr):
    A_t = np.transpose(A)
    # Loop:
    for num in range(0, numItr):
        # Hypothesis:
        y_    =  np.dot(A,x)
        # Loss function
        loss  = y_ - y
        # Display
        #print 'Gradient Step:', num
        # Gradient of A
        gradient  = np.dot(A_t,loss)
        # update
        x = x - (alpha*gradient)
    return x
```

```python
# Fiding m and c through gradient Descent.
# Converting to the form Ax = b
# Values of alpha and number of interations have been found through experiments.

ones = np.ones(len(x1))
x = [[x1],[ones]]
x = np.array(x)
x = np.transpose(x)

theta = np.zeros([2,1])
alpha = 0.0005

x = np.resize(x, [15,2])
y = np.resize(y, [15,1])

print ('Please wait Until Training is completed!')
print ('Training................................')
theta = gradientDescent(x,y, theta, alpha, 2500)
print 'Training Complete:'
print 'm = ', round(theta[0], 2)
print 'c = ', round(theta[1], 1)
```

```python
# Plotting:
m1 = round(theta[0], 2)
c1 = round(theta[1], 1)
y1 = m1*x1 + c1

data = plt.plot(x1,y, 'ro')
fitted = plt.plot(x1,y1)
plt.xlabel('x')
plt.ylabel('y = mx + c')
plt.show()
```

```python
# Normal Equation:
def line_least_squares(x,y):
    # cost function:
    x_t = np.transpose(x)
    inv = np.linalg.inv(np.dot(x_t,x))
    theta = np.dot(inv, np.dot(x_t,y))
    return theta
```

```python
# Fiding m and c through Normal Equation.

print ('Please wait Until Training is completed!')
print ('Training................................')
NE_theta = line_least_squares(x,y)
print 'Training Complete:'
print 'm = ', NE_theta[0]
print 'c = ', NE_theta[1]
```

```python
# Plotting:
m2 = NE_theta[0]
c2 = NE_theta[1]
y2 = m2*x1 + c2

data = plt.plot(x1,y, 'ro')
fitted = plt.plot(x1,y2)
plt.xlabel('x')
plt.ylabel('y = mx + c')
plt.show()
```
